```python
#pie
labels = 'Anupam', 'Faraz', 'Ketan', 'Kritik'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Faraz') (segregate)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# Graphical Analysis

```python
# Plotting a bar graph
dataset.quality.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Number of wine by quality")
plt.ylabel("Number of wine")
plt.xlabel("Quality");
```

```python
dataset.quality.value_counts()
```

```python
sns.boxplot(x=dataset["residual sugar"])
```

```python
plt.figure(figsize=(20,10))
c= dataset.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
plt.xlim(0.0,0.75)                                                      #rescaling
plt.scatter(dataset["volatile acidity"], dataset["citric acid"])
plt.xlabel("volatile acidity")
plt.ylabel("citric acid")
plt.show()
```

```python
# line graph

plt.plot(dataset["volatile acidity"][2:8])
plt.ylabel('value  of volatile acidity')
plt.xlabel("volatile acidity")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = dataset.quality.unique()
sizes = dataset.quality.value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

```python
#Histogram plot
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(dataset["citric acid"],bins= 20)
#plt.show()
```
